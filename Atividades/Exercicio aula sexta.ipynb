## Exercício para casa 2
* Criação da tabela Pessoa que permitar mudar o nome de uma cidade nas tabelas Pessoa e Cidade com um único comando

```sql
%defaultDatasource jdbc:h2:mem:dB
```

```sql
DROP TABLE IF EXISTS Pessoa;
DROP TABLE IF EXISTS Cidade;
```

```sql
CREATE TABLE Cidade (
Nome_cidade VARCHAR(30) NOT NULL,
Sigla_estado VARCHAR(2) NOT NULL,
PRIMARY KEY(Nome_cidade)
);

CREATE TABLE Pessoa (
Nome VARCHAR (100) NOT NULL,
Nome_da_mãe VARCHAR (100) NOT NULL,
Ano_nascimento INTEGER NOT NULL,
Nome_cidade_natal VARCHAR (30),
PRIMARY KEY(Nome),
FOREIGN KEY(Nome_cidade_natal)
REFERENCES Cidade(Nome_cidade)
    ON DELETE NO ACTION
    ON UPDATE CASCADE
);
```

## Exercício para casa 1 item a
* Quantos vírus trata cada medicamento

```sql
SELECT medicamento.nomeVenda, COUNT(*)
FROM medicamento, virus
WHERE virus.nomeVendaMedicamento = medicamento.nomeVenda
GROUP BY medicamento.nomeVenda;
```

## Exercício para casa 3
* Retornar o número de pessoas em cada estado

```sql
SELECT cidade.sigla_estado, COUNT(*)
FROM cidade, pessoa
WHERE pessoa.nome_cidade_natal = cidade.nome_cidade
GROUP BY cidade.sigla_estado;
```

## Exercício para casa 4 
* Qual vírus exige maior dosagem de medicamento para tipoPaciente cuja descrição é ‘idoso’

```sql
SELECT virus.nomeCientifico, trata.dosagem
FROM vírus, trata, tipoPaciente
WHERE trata.nomeCientificoVirus = virus.nomeCientifico AND trata.idTipoPaciente = tipoPaciente.id
GROUP BY trata.dosagem
HAVING paciente.descricao = 'idoso'
ORDER BY trata.dosagem DESC
LIMIT 1
```

## Exercício para casa 5 item a
* Quais compostos ativos tratam vírus com período de incubação maior que 5 dias

```sql
SELECT medicamento.compostoAtivo
FROM medicamento, vírus, trata
WHERE medicamento.nomeVenda = trata.nomeVendaMedicamento  AND vírus.nomeCientifico = trata.nomeCientificoVirus
GROUP BY medicamento.compostoAtivo
HAVING vírus.incubacao > '5'
```

## Exercício para casa 5 item b
* Quais compostos ativos tratam mais que 5 vírus

```sql
SELECT medicamento.compostoAtivo, COUNT(*) NUMERO
FROM medicamento, trata
WHERE medicamento.nomeVenda = trata.nomeVendaMedicamento
GROUP BY medicamento.compostoAtivo;
HAVING NUMERO > '5'
```
